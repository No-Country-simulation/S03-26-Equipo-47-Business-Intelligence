### <font color='lightgreen'>Librerías</font>

```python
import pandas as pd
from pathlib import Path
```

### <font color='lightgreen'>Configuración</font>

```python
# Configuración
BASE_URL = "https://raw.githubusercontent.com/No-Country-simulation/S03-26-Equipo-47-Business-Intelligence/development/data/raw"
DATA_PROCESSED = "../data/loaded"

# Crear carpeta de salida
Path(DATA_PROCESSED).mkdir(parents=True, exist_ok=True)

# Lista de archivos
archivos = [
    "compras_proveedor.csv",
    "consumo_energetico.csv",
    "encuestas_clima.csv",
    "eventos_rrhh.csv",
    "personal_nomina.csv",
    "residuos_reciclaje.csv",
    "ventas_transacciones.csv",
]

datos = {}
```

### <font color='lightgreen'>Carga de datos</font>

#### <font color='lightgreen'>> 1.1 Compras Proveedor</font>

```python
print("\n📁 compras_proveedor.csv")
try:
    url = f"{BASE_URL}/compras_proveedor.csv"
    df = pd.read_csv(
        url,
        encoding='utf-8',
        on_bad_lines='skip',
        engine='python'
    )
    if 'fecha_emision' in df.columns:
        df['fecha_emision'] = pd.to_datetime(df['fecha_emision'], errors='coerce')
    if 'fecha_pago' in df.columns:
        df['fecha_pago'] = pd.to_datetime(df['fecha_pago'], errors='coerce')
    datos['compras'] = df
    print(f"   ✅ {len(df)} registros")
except Exception as e:
    print(f"   ❌ {e}")
```

#### <font color='lightgreen'>> 1.2 Consumo Energético</font>

```python
print("\n📁 consumo_energetico.csv")
try:
    url = f"{BASE_URL}/consumo_energetico.csv"
    df = pd.read_csv(url, encoding='utf-8', on_bad_lines='skip', engine='python')
    if 'fecha' in df.columns:
        df['fecha'] = pd.to_datetime(df['fecha'], errors='coerce')
    if 'kwh_consumidos' in df.columns and 'm3_gas' in df.columns:
        df['kwh_totales'] = df['kwh_consumidos'].fillna(0) + (df['m3_gas'].fillna(0) * 10.69)
    datos['consumo_energetico'] = df
    print(f"   ✅ {len(df)} registros")
except Exception as e:
    print(f"   ❌ {e}")
```

#### <font color='lightgreen'>> 1.3 Encuesta Clima</font>

```python
print("\n📁 encuestas_clima.csv")
try:
    url = f"{BASE_URL}/encuestas_clima.csv"
    df = pd.read_csv(url, encoding='utf-8', on_bad_lines='skip', engine='python')
    if 'fecha_encuesta' in df.columns:
        df['fecha_encuesta'] = pd.to_datetime(df['fecha_encuesta'], errors='coerce')
    datos['encuestas'] = df
    print(f"   ✅ {len(df)} registros")
except Exception as e:
    print(f"   ❌ {e}")
```

#### <font color='lightgreen'>> 1.4 Eventos RRHH</font>

```python
print("\n📁 eventos_rrhh.csv")
try:
    url = f"{BASE_URL}/eventos_rrhh.csv"
    df = pd.read_csv(url, encoding='utf-8', on_bad_lines='skip', engine='python')
    if 'fecha' in df.columns:
        df['fecha'] = pd.to_datetime(df['fecha'], errors='coerce')
    datos['eventos_rrhh'] = df
    print(f"   ✅ {len(df)} registros")
except Exception as e:
    print(f"   ❌ {e}")
```

#### <font color='lightgreen'>> 1.5 Personal Nómina (formato en comillas)</font>

```python
print("\n📁 personal_nomina.csv")
try:
    url = f"{BASE_URL}/personal_nomina.csv"
    df = pd.read_csv(
        url,
        encoding='utf-8',
        quoting=1,
        doublequote=True,
        on_bad_lines='skip',
        engine='python'
    )
    for col in df.columns:
        if df[col].dtype == 'object':
            df[col] = df[col].astype(str).str.replace('"', '', regex=False)
    if 'mes' in df.columns:
        df['mes'] = pd.to_datetime(df['mes'], errors='coerce')
    if 'fecha_ingreso' in df.columns:
        df['fecha_ingreso'] = pd.to_datetime(df['fecha_ingreso'], errors='coerce')
    datos['personal_nomina'] = df
    print(f"   ✅ {len(df)} registros")
except Exception as e:
    print(f"   ❌ {e}")
```

#### <font color='lightgreen'>> 1.6 Residuos Reciclaje</font>

```python
print("\n📁 residuos_reciclaje.csv")
try:
    url = f"{BASE_URL}/residuos_reciclaje.csv"
    df = pd.read_csv(url, encoding='utf-8', on_bad_lines='skip', engine='python')
    if 'fecha_semana' in df.columns:
        df['fecha_semana'] = pd.to_datetime(df['fecha_semana'], errors='coerce')
    if 'kg_reciclados' in df.columns and 'kg_generados' in df.columns:
        df['tasa_reciclaje'] = (df['kg_reciclados'] / df['kg_generados']) * 100
    datos['residuos'] = df
    print(f"   ✅ {len(df)} registros")
except Exception as e:
    print(f"   ❌ {e}")
```

#### <font color='lightgreen'>> 1.7 Ventas Transacciones</font>

```python
print("\n📁 ventas_transacciones.csv")
try:
    url = f"{BASE_URL}/ventas_transacciones.csv"
    df = pd.read_csv(url, encoding='utf-8', on_bad_lines='skip', engine='python')
    if 'fecha' in df.columns:
        df['fecha'] = pd.to_datetime(df['fecha'], errors='coerce')
    datos['ventas'] = df
    print(f"   ✅ {len(df)} registros")
except Exception as e:
    print(f"   ❌ {e}")
```

### <font color='lightgreen'>Guardar Archivos</font>

```python
print("\n" + "=" * 50)
print("GUARDANDO ARCHIVOS...")
print("=" * 50)

for nombre, df in datos.items():
    output_file = f"{DATA_PROCESSED}/{nombre}_raw.csv"
    df.to_csv(output_file, index=False, encoding='utf-8')
    print(f"   💾 {nombre}_raw.csv ({len(df)} filas)")
```

### <font color='lightgreen'>Resumen</font>

```python
print("\n" + "=" * 50)
print("✅ CARGA COMPLETADA")
print("=" * 50)
print(f"\n📊 Dataframes cargados:")
for nombre, df in datos.items():
    print(f"   • {nombre}: {len(df):,} filas, {len(df.columns)} columnas")

print(f"\n📁 Archivos guardados en: {DATA_PROCESSED}/")
```
